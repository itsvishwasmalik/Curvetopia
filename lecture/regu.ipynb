```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('isolated_plot.png')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

b,g,r = cv2.split(image_rgb)

min_value = 0
max_value = 1
norm_type = cv2.NORM_MINMAX

b_normalized = cv2.normalize(b.astype('float'), None, min_value, max_value, norm_type)
g_normalized = cv2.normalize(g.astype('float'), None, min_value, max_value, norm_type)
r_normalized = cv2.normalize(r.astype('float'), None, min_value, max_value, norm_type)

# merge the three channels into one image
normalized_img = cv2.merge((b_normalized, g_normalized, r_normalized))

# display the original and the normalized image
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].imshow(image_rgb)
axes[0].set_title('Original Image')
# axes[0].set_xticklabels([])
# axes[0].set_yticklabels([])

axes[1].imshow(normalized_img)
axes[1].set_title('Normalised Image')
# axes[1].set_xticklabels([])
# axes[1].set_yticklabels([])

plt.show()


```

```python
# check the difference between the original and the normalized image
diff = np.abs(image_rgb - normalized_img)
plt.imshow(diff)
plt.title('Difference Image')
```
